```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from seisflows.seistools.susignal import get_wiener_filter
from seisflows.tools.graphics import _convert_to_array, plot_obs_section
from obspy import read
```

```python
def get_residuals(s, d):
    r = s.copy()
    n = len(r)
    for i in range(n):
        r[i].data[:] -= d[i].data[:]
        r[i].data = r[i].data.astype(dtype='float32')
    return r
```

```python
def get_adjoint_source(T, f, s, w):
    n = len(w)
    mu = 1e-3
    #constrct diagonal rescaling matrix
    wnorm = np.linalg.norm(w)
    R = (2 * f * np.eye(n) - T**2) / wnorm
    rw = np.matmul(R, w)
    
    # deconvolve autocorrelation of predicted from reweighted filter
    W = np.fft.rfft(w)
    P = np.fft.rfft(s)
    AC = P * np.conj(P) + mu
    
    A = np.fft.rfft()
  
    
    
```

```python
# read data and synthetics
d = read('Uz_data.su', format='SU', byteorder='<')
s = read('Uz_syn.su', format='SU', byteorder='<')
r = get_residuals(s, d)
dt = 1.25e-3
t = np.arange(0, 6, dt)

# plot output
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plot_obs_section(d, clip=1)
plt.subplot(1, 3, 2)
plot_obs_section(s, clip=1)
plt.subplot(1, 3, 3)
plot_obs_section(r, clip=10)
```

```python
i = 150;
istart, iend = (500, 3000)
td = d[i].data
ts = s[i].data
plt.plot(t[istart:iend], td[istart:iend] / abs(td[istart:iend]).max(), 'b')
plt.plot(t[istart:iend], ts[istart:iend] / abs(ts[istart:iend]).max(), 'r')
```

```python
mu = 1e-3
w = get_wiener_filter(td, ts, mu)
plt.plot(w)
sw = np.fft.irfft(np.fft.rfft(ts) * np.fft.rfft(w))
```
